```python
def get_pops(data, n_qubits, connectivity,update_rule):
    #dimension 0 is each trial
    #dimension 1 is each time step
    #dimension 2 is each qubit
    result = []
    for trial in data[f'{n_qubits} qubits'][f'{connectivity} connectivity']['unitary energy subspace 1']:
        seed = trial.split(' ')[-1]
        dat = dict(data[f'{n_qubits} qubits'][f'{connectivity} connectivity']['unitary energy subspace 1'][f'unitary seed {seed}'][f'ordering seed {update_rule}']['pops'])
        dat = {int(k.split('(')[0]): dat[k] for k in dat}
        dat = np.array([np.array([dat[k][subkey][()] for subkey in sorted(dat[k])]) for k in sorted(dat)])
        result.append(dat)
    return(np.array(result))

def get_2_qbit_dms(data, n_qubits, connectivity,update_rule):
    basis = ket.canonical_basis(2)
    #dimension 0 is each trial
    #dimension 1 is each time step (recall that sampling step is 5 by default)
    #dimension 2 is qubit pair
    #dimension 3 and 4 is the 2 qubit density matrix
    result = []
    
    def to_tuple(string):
        tuple_elements = string.strip('()').split(',')
        return tuple(int(elem.strip()) for elem in tuple_elements)

    for trial in data[f'{n_qubits} qubits'][f'{connectivity} connectivity']['unitary energy subspace 1']:
        seed = trial.split(' ')[-1]
        dat = dict(data[f'{n_qubits} qubits'][f'{connectivity} connectivity']['unitary energy subspace 1'][f'unitary seed {seed}'][f'ordering seed {update_rule}']['two_qubit_dms'])
        dat = {int(k.split('(')[0]): dat[k] for k in dat}
        
        dat = np.array([{to_tuple(subkey):DM.DensityMatrix(dat[k][subkey],basis) for subkey in sorted(dat[k])} for k in sorted(dat)])
        result.append(dat)
    return(np.array(result))
```

```python
def full_sim_tz_NP(pops,denom):
    tz_per_trial_NP_cond = []
    for pop in pops:
        NP_list=[]
        for timestep in CPcond(abs_tz(pop, denom)):
            timesteplist = []
            for ele in timestep:
                if ele <= 1:
                    ele = 0
                timesteplist.append(ele)
            NP_list.append(timesteplist)
            np.array(NP_list)
        tz_per_trial_NP_cond.append(NP_list)
    return np.array(tz_per_trial_NP_cond)

def full_sim_tz_NP_prob(pops,denom):
    tz_per_trial_NP_cond = []
    for pop in pops:
        NP_list=[]
        for timestep in CPcond(abs_tz(pop, denom)):
            timesteplist = []
            for ele in timestep:
                if ele <= 1:
                    ele = 0
                if ele > 1:
                    ele = 1
                timesteplist.append(ele)
            NP_list.append(timesteplist)
            np.array(NP_list)
        tz_per_trial_NP_cond.append(NP_list)
    return np.array(tz_per_trial_NP_cond)



def full_sim_tz(pops,denom):
    tz_per_trial= []
    for pop in pops:
        tz_per_trial.append(tz_two_q_dyns_map(pop, denom))
    return np.array(tz_per_trial)

steps=499
# insert the angle denominator chosen for Haar Q unitary eg pi/15 then denom =15
denom = 15 
#find the the numerical value of shift per qubit map in one trial index i.e. one full time series of evolution
def tz_two_q_dyns_map(pops_trial_index,denom): 
    tz = []
    np.array(pops_trial_index)
    for step_index in range(steps):
        tzi =  np.sin(np.pi/denom)*np.sin(np.pi/denom) -2*(pops_trial_index[step_index + 1 ] - np.cos(np.pi / denom)*np.cos(np.pi / denom)* pops_trial_index[step_index])
        tz.append(tzi)
    return tz

ytherm6= (5*0.1 + 0.4)/6
ytherm8= (7*0.1 + 0.4)/8
ytherm10=(9*0.1 + 0.4)/10
ytherm12=(11*0.1 + 0.4)/12

#CPcond()
#Note if this is broke the map is not only NCP also not P
def abs_tz(pops,denom):
    abs_tz=[]
    for step in tz_two_q_dyns_map(pops,denom):
        abs_step = [abs(ele) for ele in step]
        abs_tz.append(abs_step) 
    return abs_tz
def CPcond(abs_tz_pops) :
    cp_cond=[]
    for step in abs_tz_pops:
        cp_step = [ele + np.cos(np.pi / denom) * np. cos(np.pi / denom)  for ele in step]
        cp_cond.append(cp_step) 
    return cp_cond
```

```python
def time_averaged_one_point_measures_at_t(dataset, t, n):
    # Assuming dataset is list of time series data
    # Calculate the time average using window size n
    # Make sure we don't go below index 0
    start_idx = max(0, t-n)
    sum_till_t = np.sum(dataset[start_idx:t], axis=0)
    # The actual window size might be smaller if t < n
    actual_window_size = t - start_idx
    # Avoid division by zero
    if actual_window_size == 0:
        return np.zeros_like(dataset[0])
    time_average_t = (1/actual_window_size) * sum_till_t
    return np.array(time_average_t)
    
def time_averaged_one_point_measures_matrix_full_sim(dataset, n):
    # Calculate time averages for all time points with window size n
    time_average = []
    for t in range(len(dataset)):
        time_average.append(time_averaged_one_point_measures_at_t(dataset, t, n))
    return time_average

def create_NCP_prob_visualization(datas_all_ics, titles, denom):
    """
    Create visualization for NCP probability across multiple datasets,
    showing both raw data (lower alpha) and time-averaged data.
    
    Parameters:
    -----------
    datas_all_ics : list of lists
        Each inner list contains 5 datasets for one initial condition
    titles : list
        Titles for each subplot (one per initial condition)
    denom : float
        Denominator value for normalization
    
    Returns:
    --------
    fig : matplotlib.figure.Figure
        The created figure object
    """
    import matplotlib.pyplot as plt
    import numpy as np
    from matplotlib.lines import Line2D
    
    colors = ['#DC143C', '#228B22', '#FF8C00', '#8A2BE2','#8A2BE2']
    markers = ['o']
    t = np.arange(499)  # Create a list of time steps [0, 1, 2, ..., 498]
    num_datasets = len(datas_all_ics)
    print(f"Number of initial conditions: {num_datasets}")
    
    # Create figure and axes - column layout (num_datasets rows, 1 column)
    fig, axes = plt.subplots(num_datasets, 1, figsize=(12, 5))
    
    # If only one dataset, axes won't be an array
    if num_datasets == 1:
        axes = [axes]
    
    # Find global y-axis limits for consistent scaling across subplots
    global_ymax = 0
    for dataset_group in datas_all_ics:
        for data in dataset_group:
            # Check original data maximum
            original_max = np.max(data)
            if original_max > global_ymax:
                global_ymax = original_max
            
            # Also check time-averaged data maximum
            time_avg_data = time_averaged_one_point_measures_matrix_full_sim(data, 20)
            avg_max = np.max(time_avg_data)
            if avg_max > global_ymax:
                global_ymax = avg_max
    
    global_ymax = global_ymax + 0.05 * global_ymax  # Add 5% padding
    global_ymin = 0
    
    # List to store line objects for the first legend (R1-R5)
    line_objects = []
    
    # Process each initial condition dataset
    for i, dataset_group in enumerate(datas_all_ics):
        subplot_lines = []  # Store lines for this subplot
        
        # Process each of the 5 data types within this initial condition
        for j, data in enumerate(dataset_group):
            # Store original data for raw plot
            original_data = np.array(data)
            original_transposed = original_data.T
            
            # Plot the raw data with lower alpha
            axes[i].plot(t, original_transposed, color=colors[j], alpha=0.2)
            
            # Time average the data
            avg_data = time_averaged_one_point_measures_matrix_full_sim(data, 20)
            avg_data = np.array(avg_data)
            
            # Transpose data for plotting time series
            avg_data_transposed = avg_data.T
            
            # Plot the time-averaged data and store the line object
            line, = axes[i].plot(t, avg_data_transposed, label=f'N={6+2*j}', color=colors[j], alpha=0.7)
            
            # For the first subplot, collect line objects for the legend
            if i == 0:
                line_objects.append(line)
                
            # Store the line for this subplot's local legend
            subplot_lines.append(line)
        
        # Set y-axis limits the same for all subplots
        axes[i].set_ylim(global_ymin, global_ymax)
        
        # Add vertical line at t=10
        axes[i].axvline(x=10, color='red', linestyle='--', alpha=0.5)
        
        # Set axis labels - only bottom subplot gets x-label now
        if i == num_datasets - 1:
            axes[i].set_xlabel('t', fontsize=14)
        
        # All subplots get y-label in column layout
        axes[i].set_ylabel('Probability of NCP map', fontsize=14)
        
        # Create the CS legend (control strategy legend)
        legend_name = ["CS1"]
        cs_legend = Line2D([0], [0], marker=markers[i], color='gray', 
                           linestyle='-', linewidth=2, markersize=10, 
                           label=legend_name[i])
        
        # Add the CS legend to each subplot (except the first one, which is handled separately)
        if i < 1:
            axes[i].legend(handles=[cs_legend], fontsize=12, loc="upper right")
        
        # Add grid
        axes[i].grid(True, alpha=0.5)
    
    # Add both legends to the last subplot
    if line_objects:
        # First add the R1-R5 legend
        r_legend = axes[0].legend(handles=line_objects, fontsize=15, loc="lower right", 
                                 title="")
        # Add the R legend as a separate artist to the axes
        axes[0].add_artist(r_legend)
        
        # Now create and add the CS legend for the first subplot
        legend_name = ["CS1", "CS2", "CS3","CSP"]
        cs_legend = Line2D([0], [0], marker=markers[0], color='gray', 
                           linestyle='-', linewidth=2, markersize=10, 
                           label=legend_name[0])
        axes[0].legend(handles=[cs_legend], fontsize=12, loc="upper right")
    
    # Adjust layout
    plt.tight_layout()
    plt.savefig("probability_of_NCP_maps_CS123P_all_rules_12Q_column.png", dpi=300, bbox_inches='tight')
    
    return fig
```

```python
def create_system_size_vs_late_time_plot(datas_all_ics, t_range=450):
    """
    Create a scatter plot showing system size (N) on x-axis and late-time values on y-axis,
    with a best-fit curve.
    
    Parameters:
    -----------
    datas_all_ics : list of lists
        Each inner list contains 5 datasets for one initial condition
        where each dataset corresponds to a different system size N
    t_range : int, optional
        Number of time steps to average over for late-time value
        
    Returns:
    --------
    fig : matplotlib.figure.Figure
        The created figure object
    """
    import matplotlib.pyplot as plt
    import numpy as np
    from scipy.optimize import curve_fit
    
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # System sizes N = [6, 8, 10, 12, 14]
    system_sizes = [6 + 2*j for j in range(4)]
    markers = ['o', 's', '^', 'D', 'x']  # Different markers for different initial conditions
    colors = ['#DC143C', '#228B22', '#FF8C00', '#8A2BE2', '#4682B4']  # Different colors for ICs
    
    all_sizes = []
    all_late_values = []
    
    # Process each initial condition
    for i, dataset_group in enumerate(datas_all_ics):
        late_time_values = []
        
        # Process each system size within this initial condition
        for j, data in enumerate(dataset_group):
            # Convert to numpy array
            data_array = np.array(data)
            
            # Calculate late-time average (last t_range time steps)
            # Assuming data is arranged as [time_steps, runs] or similar
            late_time_data = data_array[-t_range:]
            late_time_avg = np.mean(late_time_data)
            
            # Calculate standard error for error bars
            late_time_std = np.std(late_time_data)
            late_time_se = late_time_std / np.sqrt(len(late_time_data))
            
            # Store the value
            late_time_values.append(late_time_avg)
            
            # Store for global fitting
            all_sizes.append(system_sizes[j])
            all_late_values.append(late_time_avg)
        
        # Plot this initial condition's data points with error bars
        ax.errorbar(system_sizes, late_time_values, 
                   yerr=late_time_se if len(late_time_data) > 1 else None,
                   marker=markers[i % len(markers)], 
                   color=colors[i % len(colors)],
                   label=f'CS{i+1}', 
                   markersize=10,
                   capsize=5,
                   elinewidth=2,
                   alpha=0.7)
    
    # Convert to numpy arrays for fitting
    all_sizes = np.array(all_sizes)
    all_late_values = np.array(all_late_values)
    
    # Define fitting functions to try
    def poly_func(x, a, b, c):
        return a * x**2 + b * x + c
        
    def exp_func(x, a, b, c):
        return a * np.exp(b * x) + c
    
    def power_law(x, a, b, c):
        return a * (x**b) + c
    
    def log_func(x, a, b, c):
        return a * np.log(b * x) + c
    
    def linear_func(x, a, b):
        return a * x + b
    
    # Dictionary to store all fit results
    fit_results = {}
    fit_lines = []
    
    # Try all fitting functions
    fit_functions = {
        'Quadratic': (poly_func, ['a', 'b', 'c']),
        'Exponential': (exp_func, ['a', 'b', 'c']),
        'Power Law': (power_law, ['a', 'b', 'c']),
        'Logarithmic': (log_func, ['a', 'b', 'c']),
        'Linear': (linear_func, ['a', 'b'])
    }
    
    # Color scheme for different fit types
    fit_colors = {
        'Quadratic': 'k-',
        'Exponential': 'r--',
        'Power Law': 'g-.',
        'Logarithmic': 'b:',
        'Linear': 'm-'
    }
    
    # Create smooth x values for plotting
    x_smooth = np.linspace(min(system_sizes) - 0.5, max(system_sizes) + 0.5, 100)
    
    # Try each fitting function
    for fit_name, (func, param_names) in fit_functions.items():
        try:
            # Set different initial params and bounds based on function type
            if fit_name == 'Exponential':
                p0 = [
                    np.max(all_late_values) - np.min(all_late_values),  # amplitude
                    0.1,                                                # growth rate
                    np.min(all_late_values)                             # offset
                ]
                bounds = (
                    [0, -1, -np.inf],    # Lower bounds
                    [np.inf, 1, np.inf]  # Upper bounds
                )
                params, _ = curve_fit(func, all_sizes, all_late_values, p0=p0, maxfev=10000, bounds=bounds)
            elif fit_name == 'Logarithmic':
                # For log function, b must be positive
                p0 = [1, 1, 0]
                bounds = (
                    [-np.inf, 0.001, -np.inf],  # Lower bounds
                    [np.inf, np.inf, np.inf]    # Upper bounds
                )
                params, _ = curve_fit(func, all_sizes, all_late_values, p0=p0, maxfev=10000, bounds=bounds)
            elif fit_name == 'Power Law':
                # For power law, ensure reasonable bounds
                p0 = [1, 1, 0]
                bounds = (
                    [-np.inf, -5, -np.inf],  # Lower bounds
                    [np.inf, 5, np.inf]      # Upper bounds
                )
                params, _ = curve_fit(func, all_sizes, all_late_values, p0=p0, maxfev=10000, bounds=bounds)
            else:
                # Default params for other functions
                params, _ = curve_fit(func, all_sizes, all_late_values, maxfev=10000)
            
            # Calculate fit quality metrics
            y_pred = func(np.array(all_sizes), *params)
            rss = np.sum((all_late_values - y_pred)**2)
            ss_tot = np.sum((all_late_values - np.mean(all_late_values))**2)
            r_squared = 1 - (rss / ss_tot)
            
            # Calculate adjusted R which penalizes for number of parameters
            n = len(all_sizes)  # Number of data points
            p = len(param_names)  # Number of parameters
            adj_r_squared = 1 - (1 - r_squared) * (n - 1) / (n - p - 1) if n > p + 1 else 0
            
            # Calculate AIC (Akaike Information Criterion)
            # Lower AIC indicates better model
            aic = n * np.log(rss/n) + 2 * p
            
            # Calculate BIC (Bayesian Information Criterion)
            # BIC penalizes extra parameters more strongly than AIC
            bic = n * np.log(rss/n) + p * np.log(n)
            
            # Store the results
            fit_results[fit_name] = {
                'params': params,
                'r_squared': r_squared,
                'adj_r_squared': adj_r_squared,
                'rss': rss,
                'aic': aic,
                'bic': bic,
                'func': func,
                'param_names': param_names
            }
            
            # Generate y values for smooth curve
            y_smooth = func(x_smooth, *params)
            
            # Create parameter string for label
            param_str = ''
            if fit_name == 'Quadratic':
                param_str = f"{params[0]:.3f}x + {params[1]:.3f}x + {params[2]:.3f}"
            elif fit_name == 'Exponential':
                param_str = f"{params[0]:.3f}e^({params[1]:.3f}x) + {params[2]:.3f}"
            elif fit_name == 'Power Law':
                param_str = f"{params[0]:.3f}x^{params[1]:.3f} + {params[2]:.3f}"
            elif fit_name == 'Logarithmic':
                param_str = f"{params[0]:.3f}ln({params[1]:.3f}x) + {params[2]:.3f}"
            elif fit_name == 'Linear':
                param_str = f"{params[0]:.3f}x + {params[1]:.3f}"
            
            # Add fit line to plot without using ax.lines.pop() which caused the error
            line, = ax.plot(x_smooth, y_smooth, fit_colors[fit_name], 
                           label=f'{fit_name}: {param_str}, R={r_squared:.3f}', 
                           linewidth=2, alpha=0.7)
            fit_lines.append(line)
            
            print(f"{fit_name} fit: R={r_squared:.4f}, Adj-R={adj_r_squared:.4f}, RSS={rss:.6f}, AIC={aic:.2f}, BIC={bic:.2f}")
            
        except Exception as e:
            print(f"{fit_name} fit failed: {e}")
    
    # Find the best fit according to different metrics
    if fit_results:
        # Best by R-squared
        best_r2 = max(fit_results.items(), key=lambda x: x[1]['r_squared'])
        print(f"\nBest fit by R: {best_r2[0]} (R={best_r2[1]['r_squared']:.4f})")
        
        # Best by Adjusted R-squared
        best_adj_r2 = max(fit_results.items(), key=lambda x: x[1]['adj_r_squared'])
        print(f"Best fit by Adjusted R: {best_adj_r2[0]} (Adj-R={best_adj_r2[1]['adj_r_squared']:.4f})")
        
        # Best by AIC (lower is better)
        best_aic = min(fit_results.items(), key=lambda x: x[1]['aic'])
        print(f"Best fit by AIC: {best_aic[0]} (AIC={best_aic[1]['aic']:.2f})")
        
        # Best by BIC (lower is better)
        best_bic = min(fit_results.items(), key=lambda x: x[1]['bic'])
        print(f"Best fit by BIC: {best_bic[0]} (BIC={best_bic[1]['bic']:.2f})")
    else:
        print("All fits failed. Showing only data points.")
    
    # This section is replaced by the comprehensive fitting code above
    
    # Add grid, labels, and legend
    ax.grid(True, alpha=0.3)
    ax.set_xlabel('System Size (N)', fontsize=14)
    ax.set_ylabel('Late-Time Average Probability', fontsize=14)
    ax.set_title('System Size vs. Late-Time Probability of NCP Map', fontsize=16)
    
    # Add xticks for the specific system sizes
    ax.set_xticks(system_sizes)
    
    # Add legend with appropriate sizing and positioning
    if len(fit_lines) > 3:
        # Create a more compact legend for many fits
        legend = ax.legend(fontsize=10, loc='best', ncol=2)
    else:
        legend = ax.legend(fontsize=12, loc='best')
        
    # Highlight the best fit in the legend (if any fits succeeded)
    if fit_results:
        # Get the best fit by adjusted R-squared (penalizes for model complexity)
        best_fit = max(fit_results.items(), key=lambda x: x[1]['adj_r_squared'])[0]
        
        # Add an annotation about the best fit
        text_y = min(all_late_values) if best_fit in ['Exponential', 'Power Law'] else max(all_late_values)
        ax.annotate(f'Best fit: {best_fit} (by Adjusted R)', 
                   xy=(np.mean(system_sizes), text_y),
                   xytext=(0, 20 if best_fit in ['Exponential', 'Power Law'] else -20),
                   textcoords='offset points',
                   fontsize=12, fontweight='bold',
                   ha='center', va='center',
                   bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.3),
                   arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2'))
    
    plt.tight_layout()
    # Enhanced figure saving with metadata
    plt.savefig("system_size_vs_late_time_value.png", dpi=300, bbox_inches='tight')
    
    # Also save a version with transparent background for presentations
    plt.savefig("system_size_vs_late_time_value_transparent.png", dpi=300, bbox_inches='tight', transparent=True)
    
    # For publication quality, also save as PDF and SVG
    plt.savefig("system_size_vs_late_time_value.pdf", bbox_inches='tight')
    plt.savefig("system_size_vs_late_time_value.svg", bbox_inches='tight')
    
    return fig

# Example of how to use this function with your existing data:
# 
# # Assuming datas_all_ics is your existing data structure from the original function
# create_system_size_vs_late_time_plot(datas_all_ics)
```

```python
def extract_observables(two_qubit_dms, n_qubits):
    """
    Extract single-qubit z expectation values and two-qubit correlations from density matrices.
    
    Args:
        two_qubit_dms: Array of shape (n_trials, n_timesteps, n_pairs) containing dictionaries 
                       with two-qubit density matrices
        n_qubits: Number of qubits in the system
    
    Returns:
        sz: Array of shape (n_trials, n_timesteps, n_qubits) with z expectation values
        corr: Array of shape (n_trials, n_timesteps, n_qubits, n_qubits) with z-z correlations
    """
    n_trials = two_qubit_dms.shape[0]
    n_timesteps = two_qubit_dms.shape[1]
    
    # Pauli Z matrix
    pauli_z = np.array([[1, 0], [0, -1]], dtype=complex)
    pauli_z_tensor = np.kron(pauli_z, pauli_z)
    
    # Initialize arrays as complex to avoid casting warnings
    sz = np.zeros((n_trials, n_timesteps, n_qubits), dtype=complex)
    corr = np.zeros((n_trials, n_timesteps, n_qubits, n_qubits), dtype=complex)
    
    # For each trial and timestep
    for trial in range(n_trials):
        for t in range(n_timesteps):
            # Count how many times each qubit appears to average properly
            counts = np.zeros(n_qubits)
            
            # For each qubit pair
            for pair, dm in two_qubit_dms[trial, t].items():
                i, j = pair
                
                # Get density matrix as numpy array
                rho = dm.data
                
                # Calculate local observables
                sz_i = np.trace(rho @ np.kron(pauli_z, np.eye(2, dtype=complex)))
                sz_j = np.trace(rho @ np.kron(np.eye(2, dtype=complex), pauli_z))
                
                # Calculate correlation
                corr_ij = np.trace(rho @ pauli_z_tensor) - sz_i * sz_j
                
                # Update counts
                counts[i] += 1
                counts[j] += 1
                
                # Update expectation values (will average later)
                sz[trial, t, i] += sz_i
                sz[trial, t, j] += sz_j
                
                # Store correlation
                corr[trial, t, i, j] = corr_ij
                corr[trial, t, j, i] = corr_ij
            
            # Average the expectation values
            for i in range(n_qubits):
                if counts[i] > 0:
                    sz[trial, t, i] /= counts[i]
    
    # Take real parts at the end - for Hermitian observables like Pauli-Z
    # the expectation values should be real, with any imaginary component
    # coming from numerical imprecision
    return sz.real, corr.real
    
    return sz, corr
```

```python
def plot_adjacency_sum_time_evolution(adj_matrices_datasets, titles=None, colors=None):
    """
    Plot the time evolution of the total sum of adjacency matrix values,
    averaged over trials for multiple datasets.
    
    Parameters:
    -----------
    adj_matrices_datasets : list of numpy arrays
        Each element is a dataset containing adjacency matrices over time and trials.
        Expected shape for each dataset: (n_trials, n_timesteps, n_nodes, n_nodes)
    titles : list of str, optional
        Titles/labels for each dataset
    colors : list of str, optional
        Colors for plotting each dataset
    
    Returns:
    --------
    fig : matplotlib.figure.Figure
        The created figure object
    """
    # Default colors if none provided
    if colors is None:
        colors = ['#DC143C', '#228B22', '#FF8C00', '#8A2BE2', '#4682B4']
    
    # Default titles if none provided
    if titles is None:
        titles = [f'Dataset {i+1}' for i in range(len(adj_matrices_datasets))]
    
    # Create figure
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # Process each dataset
    for i, adj_matrices in enumerate(adj_matrices_datasets):
        n_trials, n_timesteps = adj_matrices.shape[0], adj_matrices.shape[1]
        
        # Calculate the sum of adjacency values at each time step for each trial
        total_sum = np.zeros((n_trials, n_timesteps))
        for trial in range(n_trials):
            for t in range(n_timesteps):
                total_sum[trial, t] = np.sum(adj_matrices[trial, t])
        
        # Calculate mean and standard error across trials
        mean_sum = np.mean(total_sum, axis=0)
        std_sum = np.std(total_sum, axis=0)
        sem_sum = std_sum / np.sqrt(n_trials)  # Standard Error of the Mean
        
        # Time steps array
        time_steps = np.arange(n_timesteps)
        
        # Plot mean with shaded error region
        ax.plot(time_steps, mean_sum, label=titles[i], color=colors[i % len(colors)], linewidth=2)
        ax.fill_between(time_steps, mean_sum - sem_sum, mean_sum + sem_sum, 
                       color=colors[i % len(colors)], alpha=0.2)
    
    # Add grid, labels, and legend
    ax.grid(True, alpha=0.3)
    ax.set_xlabel('Time Step', fontsize=14)
    ax.set_ylabel('Total Sum of Adjacency Values', fontsize=14)
    ax.set_title('Time Evolution of Adjacency Matrix Sum', fontsize=16)
    
    # Add legend
    ax.legend(fontsize=12, loc='best')
    
    plt.tight_layout()
    plt.savefig("adjacency_sum_time_evolution.png", dpi=300, bbox_inches='tight')
    
    return fig


def plot_adjacency_mean_std_time_evolution(adj_matrices_datasets, titles=None, colors=None):
    """
    Plot the time evolution of the mean and standard deviation of adjacency matrix elements,
    averaged over trials for multiple datasets.
    
    Parameters:
    -----------
    adj_matrices_datasets : list of numpy arrays
        Each element is a dataset containing adjacency matrices over time and trials.
        Expected shape for each dataset: (n_trials, n_timesteps, n_nodes, n_nodes)
    titles : list of str, optional
        Titles/labels for each dataset
    colors : list of str, optional
        Colors for plotting each dataset
    
    Returns:
    --------
    fig : matplotlib.figure.Figure
        The created figure object
    """
    # Default colors if none provided
    if colors is None:
        colors = ['#DC143C', '#228B22', '#FF8C00', '#8A2BE2', '#4682B4']
    
    # Default titles if none provided
    if titles is None:
        titles = [f'Dataset {i+1}' for i in range(len(adj_matrices_datasets))]
    
    # Create figure with two subplots
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), sharex=True)
    
    # Process each dataset
    for i, adj_matrices in enumerate(adj_matrices_datasets):
        n_trials, n_timesteps = adj_matrices.shape[0], adj_matrices.shape[1]
        
        # Arrays to store mean and std values for each trial and time step
        means = np.zeros((n_trials, n_timesteps))
        stds = np.zeros((n_trials, n_timesteps))
        
        # Calculate mean and std of adjacency values at each time step for each trial
        for trial in range(n_trials):
            for t in range(n_timesteps):
                # Flatten the adjacency matrix to get all elements
                adj_elements = adj_matrices[trial, t].flatten()
                means[trial, t] = np.mean(adj_elements)
                stds[trial, t] = np.std(adj_elements)
        
        # Calculate mean and standard error across trials
        mean_of_means = np.abs(np.mean(means, axis=0))
        std_of_means = np.std(means, axis=0)
        sem_of_means = std_of_means / np.sqrt(n_trials)
        
        mean_of_stds = np.mean(stds, axis=0)
        std_of_stds = np.std(stds, axis=0)
        sem_of_stds = std_of_stds / np.sqrt(n_trials)
        
        # Time steps array
        time_steps = np.arange(n_timesteps)
        
        # Plot mean values with shaded error region in top subplot
        ax1.plot(time_steps, mean_of_means, label=titles[i], color=colors[i % len(colors)], linewidth=2)
        ax1.fill_between(time_steps, mean_of_means - sem_of_means, mean_of_means + sem_of_means, 
                        color=colors[i % len(colors)], alpha=0.2)
        
        # Plot std values with shaded error region in bottom subplot
        ax2.plot(time_steps, mean_of_stds, label=titles[i], color=colors[i % len(colors)], linewidth=2)
        ax2.fill_between(time_steps, mean_of_stds - sem_of_stds, mean_of_stds + sem_of_stds, 
                        color=colors[i % len(colors)], alpha=0.2)
    
    # Add grid, labels, and legends
    ax1.grid(True, alpha=0.3)
    ax2.grid(True, alpha=0.3)
    
    ax1.set_ylabel('Mean of Adjacency Values', fontsize=14)
    ax1.set_title('Time Evolution of Adjacency Matrix Statistics', fontsize=16)
    ax1.legend(fontsize=12, loc='best')
    
    ax2.set_xlabel('Time Step', fontsize=14)
    ax2.set_ylabel('Std Dev of Adjacency Values', fontsize=14)
    ax2.legend(fontsize=12, loc='best')
    
    plt.tight_layout()
    plt.savefig("adjacency_mean_std_time_evolution.png", dpi=300, bbox_inches='tight')
    
    return fig



# Assuming adj_matrices_datasets is a list of your 5 datasets
# Each dataset has shape (100 trials, 500 time steps, n_nodes, n_nodes)
titles = ['Dataset 1', 'Dataset 2', 'Dataset 3', 'Dataset 4']
adj_matrices_datasets=[corr_greedy_c2_6_th[1],corr_greedy_c2_8_th[1],corr_greedy_c2_10_th[1],corr_greedy_c2_12_th[1]]
plot_adjacency_sum_time_evolution(adj_matrices_datasets, titles)
plot_adjacency_mean_std_time_evolution(adj_matrices_datasets, titles)
```

```python
def plot_late_time_adjacency_vs_system_size_single(adj_matrices_list, system_sizes, t_range=50, 
                                              color='#228B22', title='Adjacency Matrix Statistics'):
    """
    Create scatter plots showing late-time mean and standard deviation of adjacency matrices
    versus system size, with best-fit curves for a single dataset series.
    
    Parameters:
    -----------
    adj_matrices_list : list
        List of adjacency matrix datasets for different system sizes.
        Each dataset has shape: (n_trials, n_timesteps, n_nodes, n_nodes)
    system_sizes : list
        List of system sizes corresponding to each dataset (e.g., [6, 8, 10, 12, 14])
    t_range : int, optional
        Number of time steps to average over for late-time value
    color : str, optional
        Color for plotting the dataset
    title : str, optional
        Title/label for the dataset
    
    Returns:
    --------
    fig : matplotlib.figure.Figure
        The created figure object
    """
    # Create figure with two subplots side by side
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
    
    # Define fitting functions to try
    def poly_func(x, a, b, c):
        return a * x**2 + b * x + c
        
    def exp_func(x, a, b, c):
        return a * np.exp(b * x) + c
    
    def power_law(x, a, b, c):
        return a * (x**b) + c
    
    def log_func(x, a, b, c):
        return a * np.log(b * x) + c
    
    def linear_func(x, a, b):
        return a * x + b
    
    # Fitting functions to try
    fit_functions = {
        'Quadratic': (poly_func, ['a', 'b', 'c']),
        'Exponential': (exp_func, ['a', 'b', 'c']),
        'Power Law': (power_law, ['a', 'b', 'c']),
        'Logarithmic': (log_func, ['a', 'b', 'c']),
        'Linear': (linear_func, ['a', 'b'])
    }
    
    # Color scheme for different fit types
    fit_colors = {
        'Quadratic': 'k-',
        'Exponential': 'r--',
        'Power Law': 'g-.',
        'Logarithmic': 'b:',
        'Linear': 'm-'
    }
    
    # Lists to store late-time values for each system size
    late_time_means = []
    late_time_stds = []
    late_time_means_err = []
    late_time_stds_err = []
    
    # Process each system size
    for j, adj_matrices in enumerate(adj_matrices_list):
        n_trials, n_timesteps = adj_matrices.shape[0], adj_matrices.shape[1]
        
        # Arrays to store mean and std values for each trial and time step
        means = np.zeros((n_trials, n_timesteps))
        stds = np.zeros((n_trials, n_timesteps))
        
        # Calculate mean and std of adjacency values at each time step for each trial
        for trial in range(n_trials):
            for t in range(n_timesteps):
                # Flatten the adjacency matrix to get all elements
                adj_elements = adj_matrices[trial, t].flatten()
                means[trial, t] = np.abs(np.mean(adj_elements))
                stds[trial, t] = np.std(adj_elements)
        
        # Calculate late-time average (last t_range time steps)
        late_time_means_data = means[:, -t_range:]
        late_time_stds_data = stds[:, -t_range:]
        
        # Average over time and trials
        late_time_mean_avg = np.mean(late_time_means_data)
        late_time_std_avg = np.mean(late_time_stds_data)
        
        # Calculate standard error across trials for error bars
        late_time_mean_err = np.std(np.mean(late_time_means_data, axis=1)) / np.sqrt(n_trials)
        late_time_std_err = np.std(np.mean(late_time_stds_data, axis=1)) / np.sqrt(n_trials)
        
        # Store values
        late_time_means.append(late_time_mean_avg)
        late_time_stds.append(late_time_std_avg)
        late_time_means_err.append(late_time_mean_err)
        late_time_stds_err.append(late_time_std_err)
    
    # Plot data points with error bars for mean values
    ax1.errorbar(system_sizes, late_time_means, yerr=late_time_means_err,
                fmt='o', markersize=10, color=color, label=title,
                capsize=5, elinewidth=2, alpha=0.7)
    
    # Plot data points with error bars for std values
    ax2.errorbar(system_sizes, late_time_stds, yerr=late_time_stds_err,
                fmt='o', markersize=10, color=color, label=title,
                capsize=5, elinewidth=2, alpha=0.7)
    
    # Try each fitting function for mean values
    best_fit_results = {}
    
    for panel_name, (y_data, y_err) in zip(['Mean', 'Std'], [(late_time_means, late_time_means_err), 
                                            (late_time_stds, late_time_stds_err)]):
        ax = ax1 if panel_name == 'Mean' else ax2
        
        # Convert to numpy arrays for fitting
        x_data = np.array(system_sizes)
        y_data = np.array(y_data)
        
        # Create smooth x values for plotting
        x_smooth = np.linspace(min(system_sizes) - 0.5, max(system_sizes) + 0.5, 100)
        
        # Dictionary to store fit results
        fit_results = {}
        
        # Try each fitting function
        for fit_name, (func, param_names) in fit_functions.items():
            try:
                # Handle different fit function initializations
                if fit_name == 'Exponential':
                    p0 = [np.max(y_data) - np.min(y_data), 0.1, np.min(y_data)]
                    bounds = ([0, -1, -np.inf], [np.inf, 1, np.inf])
                    params, _ = curve_fit(func, x_data, y_data, p0=p0, maxfev=10000, bounds=bounds)
                    print(params)
                elif fit_name == 'Logarithmic':
                    p0 = [1, 1, 0]
                    bounds = ([-np.inf, 0.001, -np.inf], [np.inf, np.inf, np.inf])
                    params, _ = curve_fit(func, x_data, y_data, p0=p0, maxfev=10000, bounds=bounds)
                elif fit_name == 'Power Law':
                    p0 = [1, 1, 0]
                    bounds = ([-np.inf, -5, -np.inf], [np.inf, 5, np.inf])
                    params, _ = curve_fit(func, x_data, y_data, p0=p0, maxfev=10000, bounds=bounds)
                else:
                    params, _ = curve_fit(func, x_data, y_data, maxfev=10000)
                
                # Calculate fit quality metrics
                y_pred = func(x_data, *params)
                rss = np.sum((y_data - y_pred)**2)
                ss_tot = np.sum((y_data - np.mean(y_data))**2)
                r_squared = 1 - (rss / ss_tot)
                
                # Calculate adjusted R for model selection
                n = len(x_data)
                p = len(param_names)
                adj_r_squared = 1 - (1 - r_squared) * (n - 1) / (n - p - 1) if n > p + 1 else 0
                
                # Calculate AIC (Akaike Information Criterion) - lower is better
                aic = n * np.log(rss/n) + 2 * p
                
                # Calculate BIC (Bayesian Information Criterion) - lower is better
                bic = n * np.log(rss/n) + p * np.log(n)
                
                # Store results
                fit_results[fit_name] = {
                    'params': params,
                    'r_squared': r_squared,
                    'adj_r_squared': adj_r_squared,
                    'aic': aic,
                    'bic': bic,
                    'func': func
                }
                
                # Generate y values for smooth curve
                y_smooth = func(x_smooth, *params)
                
                # Create parameter string for label
                param_str = ''
                if fit_name == 'Quadratic':
                    param_str = f"{params[0]:.3f}x + {params[1]:.3f}x + {params[2]:.3f}"
                elif fit_name == 'Exponential':
                    param_str = f"{params[0]:.3f}e^({params[1]:.3f}x) + {params[2]:.3f}"
                    print(param_str)
                elif fit_name == 'Power Law':
                    param_str = f"{params[0]:.3f}x^{params[1]:.3f} + {params[2]:.3f}"
                elif fit_name == 'Logarithmic':
                    param_str = f"{params[0]:.3f}ln({params[1]:.3f}x) + {params[2]:.3f}"
                elif fit_name == 'Linear':
                    param_str = f"{params[0]:.3f}x + {params[1]:.3f}"
                
                # Plot this fit
                line, = ax.plot(x_smooth, y_smooth, fit_colors[fit_name], 
                              label=f'{fit_name}: R={r_squared:.3f}', linewidth=2, alpha=0.5)
                
                # Print fit statistics
                print(f"{panel_name} - {fit_name} fit: R={r_squared:.4f}, Adj-R={adj_r_squared:.4f}, AIC={aic:.2f}, BIC={bic:.2f}")
                
            except Exception as e:
                print(f"{panel_name} - {fit_name} fit failed: {e}")
        
        # Find the best fit according to adjusted R
        if fit_results:
            best_fit = max(fit_results.items(), key=lambda x: x[1]['adj_r_squared'])
            best_fit_name = best_fit[0]
            best_fit_data = best_fit[1]
            best_params = best_fit_data['params']
            
            # Store for later use
            best_fit_results[panel_name] = {
                'name': best_fit_name,
                'r_squared': best_fit_data['r_squared'],
                'adj_r_squared': best_fit_data['adj_r_squared'],
                'func': best_fit_data['func'],
                'params': best_params
            }
            
            # Generate best fit curve for annotation
            y_smooth = best_fit_data['func'](x_smooth, *best_params)
            
            # Add annotation about the best fit
            param_str = ''
            if best_fit_name == 'Quadratic':
                param_str = f"{best_params[0]:.3f}x + {best_params[1]:.3f}x + {best_params[2]:.3f}"
            elif best_fit_name == 'Exponential':
                param_str = f"{best_params[0]:.3f}e^({best_params[1]:.3f}x) + {best_params[2]:.3f}"
            elif best_fit_name == 'Power Law':
                param_str = f"{best_params[0]:.3f}x^{best_params[1]:.3f} + {best_params[2]:.3f}"
            elif best_fit_name == 'Logarithmic':
                param_str = f"{best_params[0]:.3f}ln({best_params[1]:.3f}x) + {best_params[2]:.3f}"
            elif best_fit_name == 'Linear':
                param_str = f"{best_params[0]:.3f}x + {best_params[1]:.3f}"
            
            # Highlight best fit
            ax.plot(x_smooth, y_smooth, color=color, linewidth=3,
                   label=f'Best fit ({best_fit_name}): R={best_fit_data["r_squared"]:.3f}')
            
            # Add text annotation for the best fit
            text_y = min(y_data) if panel_name == 'Mean' else max(y_data)
            text_offset = -30 if panel_name == 'Mean' else 20
            
            ax.annotate(f'Best fit: {best_fit_name}\n{param_str}\nR={best_fit_data["r_squared"]:.3f}',
                       xy=(np.mean(system_sizes), text_y),
                       xytext=(0, text_offset),
                       textcoords='offset points',
                       fontsize=12, color=color,
                       ha='center', va='center',
                       bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.8),
                       arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2', color=color))
    
    # Add grid, labels, and legends
    for ax in [ax1, ax2]:
        ax.grid(True, alpha=0.3)
        ax.set_xlabel('System Size (N)', fontsize=14)
        ax.set_xticks(system_sizes)
    
    ax1.set_ylabel('Late-Time Mean of Adjacency Values', fontsize=14)
    ax1.set_title('Late-Time Mean vs System Size', fontsize=16)
    
    ax2.set_ylabel('Late-Time Std Dev of Adjacency Values', fontsize=14)
    ax2.set_title('Late-Time Std Dev vs System Size', fontsize=16)
    
    # Create custom legend with only the data points and best fit line
    for ax in [ax1, ax2]:
        handles, labels = ax.get_legend_handles_labels()
        # Get only the data points and best fit line
        data_handle = handles[0]
        best_fit_handle = handles[-1]
        ax.legend([data_handle, best_fit_handle], [title, labels[-1]], fontsize=12, loc='upper left')
    
    plt.tight_layout()
    plt.savefig("late_time_adjacency_vs_system_size.png", dpi=300, bbox_inches='tight')
    
    # Also save as PDF for publication quality
    plt.savefig("late_time_adjacency_vs_system_size.pdf", bbox_inches='tight')
    
    return fig, best_fit_results



system_sizes = [6, 8, 10, 12]  # Your system sizes
titles = ['Dataset 1', 'Dataset 2', 'Dataset 3', 'Dataset 4']  # Your dataset labels

fig, best_fits = plot_late_time_adjacency_vs_system_size_single(
    adj_matrices_datasets,  # Your existing data structure
    system_sizes,
    t_range=50,             # Number of final timesteps to average
    title='Adjacency Matrix Evolution'
)

# You can access the best fit information
print("Best fit for Mean:", best_fits['Mean']['name'])
print("Best fit equation parameters:", best_fits['Mean']['params'])
```

```python
def exponential_fss_analysis(datas_all_ics, t_range=450):
    """
    Analyze system size dependence with exponential fit and finite size scaling
    
    Parameters:
    -----------
    datas_all_ics : list of lists
        Each inner list contains datasets for one initial condition
        where each dataset corresponds to a different system size N
    t_range : int, optional
        Number of time steps to average over for late-time value
        
    Returns:
    --------
    fig : matplotlib.figure.Figure
        The created figure with exponential fit and FSS analysis
    """
    import matplotlib.pyplot as plt
    import numpy as np
    from scipy.optimize import curve_fit
    
    # Create two subplots - one for direct fit, one for FSS
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))
    
    # System sizes N = [6, 8, 10, 12, 14]
    system_sizes = [6 + 2*j for j in range(5)]  # Added 14 to your list
    markers = ['o', 's', '^', 'D', 'x']  # Different markers for different initial conditions
    colors = ['#DC143C', '#228B22', '#FF8C00', '#8A2BE2', '#4682B4']
    all_sizes = []
    all_late_values = []
    
    # Process each initial condition
    for i, dataset_group in enumerate(datas_all_ics):
        late_time_values = []
        
        # Process each system size within this initial condition
        for j, data in enumerate(dataset_group):
            if j >= len(system_sizes):
                continue  # Skip if out of range
                
            # Convert to numpy array
            data_array = np.array(data)
            
            # Calculate late-time average
            late_time_data = data_array[-t_range:]
            late_time_avg = np.mean(late_time_data)
            
            # Calculate standard error for error bars
            late_time_std = np.std(late_time_data)
            late_time_se = late_time_std / np.sqrt(len(late_time_data))
            
            # Store the value
            late_time_values.append(late_time_avg)
            
            # Store for global fitting
            all_sizes.append(system_sizes[j])
            all_late_values.append(late_time_avg)
        
        # Plot this initial condition's data points with error bars
        ax1.errorbar(system_sizes[:len(late_time_values)], late_time_values, 
                   yerr=late_time_se if len(late_time_data) > 1 else None,
                   marker=markers[i % len(markers)], 
                   color="#DC143C",
                   label=f'R{i+2}', 
                   markersize=10,
                   capsize=5,
                   elinewidth=2,
                   alpha=0.7)
    
    # Convert to numpy arrays for fitting
    all_sizes = np.array(all_sizes)
    all_late_values = np.array(all_late_values)
    
    # Define exponential fitting function
    def exp_func(x, a, b, c):
        return a * np.exp(b * x) + c
    
    # Set initial parameters and bounds for exponential fit
    p0 = [
        np.max(all_late_values) - np.min(all_late_values),  # amplitude
        0.1,                                              # growth rate
        np.min(all_late_values)                           # offset
    ]
    bounds = (
        [0, -1, -np.inf],    # Lower bounds
        [np.inf, 1, np.inf]  # Upper bounds
    )
    
    # Perform the exponential fit
    try:
        params, pcov = curve_fit(exp_func, all_sizes, all_late_values, p0=p0, maxfev=10000, bounds=bounds)
        a, b, c = params
        
        # Get parameter errors from covariance matrix
        param_errors = np.sqrt(np.diag(pcov))
        
        # Calculate fit quality metrics
        y_pred = exp_func(all_sizes, *params)
        rss = np.sum((all_late_values - y_pred)**2)
        ss_tot = np.sum((all_late_values - np.mean(all_late_values))**2)
        r_squared = 1 - (rss / ss_tot)
        
        # Generate smooth curve for plotting
        x_smooth = np.linspace(min(system_sizes) - 0.5, max(system_sizes) + 0.5, 100)
        y_smooth = exp_func(x_smooth, *params)
        
        # Plot the fit curve
        line, = ax1.plot(x_smooth, y_smooth, 'r--', 
                       label=f'Exp Fit: {a:.3f}e^({b:.3f}x) + {c:.3f}, R={r_squared:.3f}', 
                       linewidth=2)
        
        print(f"Exponential fit results:")
        print(f"A = {a:.6f}  {param_errors[0]:.6f}")
        print(f"B = {b:.6f}  {param_errors[1]:.6f}")
        print(f"C = {c:.6f}  {param_errors[2]:.6f}")
        print(f"R = {r_squared:.6f}")
        
        # Extract the characteristic length scale  from the exponential fit
        # For exponential decay/growth: exp(B*x) where B = 1/ for decay or -1/ for growth
        # The sign depends on whether B is positive (growth) or negative (decay)
        xi = abs(1/b)  # Absolute value gives the characteristic length regardless of sign
        xi_error = abs(param_errors[1]/(b**2))  # Error propagation
        
        print(f"\nFinite Size Scaling Parameters:")
        print(f"Correlation length  = {xi:.6f}  {xi_error:.6f}")
        
        # Finite Size Scaling Analysis
        # Plot (O(L) - O_) versus L/ for FSS transformation
        # O_ is the asymptotic value C from our fit
        
        # Calculate scaling variable L/
        scaling_variable = all_sizes/xi
        
        # Calculate scaled observable (O(L) - O_)
        scaled_observable = all_late_values - c
        
        # For classic FSS, we might plot (O(L) - O_)L^(/) vs L/
        # where / is another critical exponent
        # Since we have an exponential form, we can use the approximation:
        # (O(L) - O_)  A*exp(-L/) = A*exp(-scaling_variable)
        
        # So our scaling function should be approximately: f(L/) = exp(-L/)
        # Let's verify this by plotting scaled_observable vs scaling_variable
        
        # Plot individual points in FSS plot
        for i, dataset_group in enumerate(datas_all_ics):
            group_sizes = []
            group_values = []
            
            for j, data in enumerate(dataset_group):
                if j >= len(system_sizes):
                    continue
                    
                group_sizes.append(system_sizes[j])
                
                # Calculate late-time average
                data_array = np.array(data)
                late_time_avg = np.mean(data_array[-t_range:])
                group_values.append(late_time_avg)
            
            # Convert to numpy arrays
            group_sizes = np.array(group_sizes)
            group_values = np.array(group_values)
            
            # Calculate scaling variables for this group
            group_scaling_var = group_sizes/xi
            group_scaled_obs = group_values - c
            
            # Plot in FSS plot
            ax2.scatter(group_scaling_var, group_scaled_obs, 
                      marker=markers[i % len(markers)], 
                      color=colors[i % len(colors)],
                      label=f'CS{i+1}, R_1', 
                      s=100,
                      alpha=0.7)
        
        # Generate theoretical scaling function curve
        x_theory = np.linspace(0, max(scaling_variable)*1.2, 100)
        y_theory = a * np.exp(-x_theory)
        ax2.plot(x_theory, y_theory, 'k-', 
                label=f'Scaling function: f(L/) = {a:.3f}e^(-L/)', 
                linewidth=2)
        
        # Calculate data collapse quality
        predicted_scaled_values = a * np.exp(-scaling_variable)
        collapse_mse = np.mean((scaled_observable - predicted_scaled_values)**2)
        collapse_r2 = 1 - np.sum((scaled_observable - predicted_scaled_values)**2) / np.sum((scaled_observable - np.mean(scaled_observable))**2)
        
        print(f"Data collapse quality: MSE = {collapse_mse:.6f}, R = {collapse_r2:.6f}")
        
        # Check if infinity value is reasonable
        o_infinity = c
        print(f"Asymptotic value (L): O_ = {o_infinity:.6f}")
        
        # Optional: if applicable, calculate critical exponent (in case of power law corrections)
        # This assumes behavior like: O(L) - O_ ~ L^(-/) * exp(-L/)
        # Would require more sophisticated analysis
        
    except Exception as e:
        print(f"Exponential fit failed: {e}")
        ax1.text(0.5, 0.5, f"Fit Failed: {str(e)}", transform=ax1.transAxes, 
               ha='center', va='center', fontsize=12, color='#DC143C')
    
    # Configure first plot (direct fit)
    ax1.grid(True, alpha=0.3)
    ax1.set_xlabel('System Size (N)', fontsize=25)
    ax1.set_ylabel(r'$P(\Lambda_{NCP})$', fontsize=25)
    #ax1.set_title('System Size vs. Late-Time Value with Exponential Fit', fontsize=14)
    ax1.set_xticks(system_sizes)
    ax1.legend(fontsize=15, loc='best')
    
    # Configure second plot (FSS analysis)
    ax2.grid(True, alpha=0.3)
    ax2.set_xlabel('Scaling Variable (L/)', fontsize=14)
    ax2.set_ylabel('Scaled Observable (O(L) - O_)', fontsize=14)
    ax2.set_title('Finite Size Scaling Analysis', fontsize=14)
    
    # Use log scale for y-axis in FSS plot for exponential behavior
    ax2.set_yscale('log')
    ax2.legend(fontsize=10, loc='best')
    
    # Add annotation about FSS theory
    ax2.annotate('For exponential behavior:\nO(L) - O_ = Ae^(-L/)', 
               xy=(0.05, 0.05), xycoords='axes fraction',
               fontsize=20, bbox=dict(boxstyle='round', fc='white', alpha=0.8))
    
    plt.tight_layout()
    plt.savefig("fss_analysis_R2.png", dpi=300, bbox_inches='tight')
    plt.savefig("fss_analysis.pdf", bbox_inches='tight')
    return fig

#Non CP count
fig = exponential_fss_analysis(datas_all_ics)
plt.show()
```

```python
def analyze_noise_scaling(all_pops, N_values):
    """
    Analyze how mean standard deviation of noise varies with dimension count (N)
    with exponential fit only.
    
    Parameters:
    -----------
    all_pops : list of arrays
        List of arrays containing data for different N values
        Each array has shape (n_trials, n_timesteps, N)
    N_values : list or array
        List of N values corresponding to each entry in all_pops
        
    Returns:
    --------
    fig : matplotlib.figure.Figure
        Figure with exponential fit plot
    """
    # Create figure with a single plot
    fig, ax = plt.subplots(figsize=(8, 7))
    
    # Calculate mean standard deviation for each N
    results = []
    for data_subset in all_pops:
        # Calculate mean across trials
        mean_subset = np.mean(data_subset, axis=0)
        
        # Calculate deviations
        deviations_subset = data_subset - mean_subset
        
        # Calculate standard deviation for each dimension
        std_per_dim = np.std(deviations_subset, axis=0)
        
        # Calculate mean standard deviation across dimensions
        mean_std = np.mean(std_per_dim, axis=1)
        
        # Store the grand mean (single value for this N)
        results.append(np.mean(mean_std))
    
    # Plot raw data points with crimson color (DC143C)
    ax.plot(N_values, results, 'o-', color='#DC143C', markersize=10, label='R2')
    
    # Define exponential fitting function
    def exp_func(x, a, b, c):
        return a * np.exp(b * x) + c
    
    # Set initial parameters for fit
    p0 = [max(results) - min(results), -0.1, min(results)]
    
    try:
        # Perform exponential fit
        params, pcov = curve_fit(exp_func, N_values, results, p0=p0, maxfev=10000)
        a, b, c = params
        
        # Get parameter errors
        param_errors = np.sqrt(np.diag(pcov))
        
        # Calculate fit quality metrics
        y_pred = exp_func(np.array(N_values), *params)
        residuals = np.array(results) - y_pred
        ss_res = np.sum(residuals**2)
        ss_tot = np.sum((np.array(results) - np.mean(results))**2)
        r_squared = 1 - (ss_res / ss_tot)
        
        # Generate smooth curve for plotting
        x_smooth = np.linspace(min(N_values) - 0.5, max(N_values) + 0.5, 100)
        y_smooth = exp_func(x_smooth, *params)
        
        # Plot the fit curve in blue
        ax.plot(x_smooth, y_smooth, '--', color='red', 
                linewidth=2, 
                label=f'Exp Fit: {a:.3f}e^({b:.3f}x) + {c:.3f}, R={r_squared:.3f}')
        
        print(f"Exponential fit results:")
        print(f"A = {a:.6f}  {param_errors[0]:.6f}")
        print(f"B = {b:.6f}  {param_errors[1]:.6f}")
        print(f"C = {c:.6f}  {param_errors[2]:.6f}")
        print(f"R = {r_squared:.6f}")
        
    except Exception as e:
        print(f"Exponential fit failed: {e}")
        ax.text(0.5, 0.5, f"Fit Failed: {str(e)}", transform=ax.transAxes, 
               ha='center', va='center', fontsize=12, color='red')
    
    # Configure plot
    ax.grid(True, alpha=0.3)
    ax.set_xlabel('N', fontsize=30)
    ax.set_ylabel(r'$\langle\sigma_{z}\rangle_{rm ens}$', fontsize=30)
    #ax.set_title('Noise Variation with Dimension Count', fontsize=14)
    ax.legend(fontsize=15, loc='best')
    
    plt.tight_layout()
    plt.savefig("noise_exp_fit_analysis_R2.png", dpi=300, bbox_inches='tight')
    
    return fig


N_values = [6, 8, 10, 12]  # Different values of N to test

# Pops std 
all_pops=[pops_greedy_6_c2,pops_greedy_8_c2,pops_greedy_10_c2,pops_greedy_12_c2]
def transform_array(all_pops):
    """
    Transform an array of shape (100, 500, N) from p to 1-2p
    
    Parameters:
    pops (numpy.ndarray): Array of shape (100, 500, N) containing probability values p
    
    Returns:
    numpy.ndarray: Transformed array with the same shape where each value p is converted to 1-2p
    """
    all_p=[]
    for pops in all_pops:
    # Create a copy to avoid modifying the original array
        transformed = 1 - 2 * pops
        all_p.append(transformed)
    
    return all_p
all_pops_t=transform_array(all_pops)
fig = analyze_noise_scaling(all_pops_t, N_values)
```

```python
def plot_adjacency_with_fss(adj_matrices_list, system_sizes, t_range=50, 
                           color='#228B22', title='Adjacency Matrix Statistics'):
    """
    Create plots showing:
    1. Late-time mean and standard deviation vs system size with exponential fits
    2. Finite size scaling analysis of the standard deviation
    
    Parameters:
    -----------
    adj_matrices_list : list
        List of adjacency matrix datasets for different system sizes.
        Each dataset has shape: (n_trials, n_timesteps, n_nodes, n_nodes)
    system_sizes : list
        List of system sizes corresponding to each dataset (e.g., [6, 8, 10, 12, 14])
    t_range : int, optional
        Number of time steps to average over for late-time value
    color : str, optional
        Color for plotting the dataset
    title : str, optional
        Title/label for the dataset
    
    Returns:
    --------
    fig : matplotlib.figure.Figure
        The created figure object
    fit_results : dict
        The exponential fit parameters and metrics
    """
    # Create figure with three subplots
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(24, 8))
    
    # Define exponential fitting function
    def exp_func(x, a, b, c):
        return a * np.exp(b * x) + c
    
    # Lists to store late-time values for each system size
    late_time_means = []
    late_time_stds = []
    late_time_means_err = []
    late_time_stds_err = []
    
    # Process each system size
    for j, adj_matrices in enumerate(adj_matrices_list):
        n_trials, n_timesteps = adj_matrices.shape[0], adj_matrices.shape[1]
        
        # Arrays to store mean and std values for each trial and time step
        means = np.zeros((n_trials, n_timesteps))
        stds = np.zeros((n_trials, n_timesteps))
        
        # Calculate mean and std of adjacency values at each time step for each trial
        for trial in range(n_trials):
            for t in range(n_timesteps):
                # Flatten the adjacency matrix to get all elements
                adj_elements = adj_matrices[trial, t].flatten()
                means[trial, t] = np.mean(np.abs(adj_elements))
                stds[trial, t] = np.std(adj_elements)
        
        # Calculate late-time average (last t_range time steps)
        late_time_means_data = means[:, -t_range:]
        late_time_stds_data = stds[:, -t_range:]
        
        # Average over time and trials
        late_time_mean_avg = np.mean(late_time_means_data)
        late_time_std_avg = np.mean(late_time_stds_data)
        
        # Calculate standard error across trials for error bars
        late_time_mean_err = np.std(np.mean(late_time_means_data, axis=1)) / np.sqrt(n_trials)
        late_time_std_err = np.std(np.mean(late_time_stds_data, axis=1)) / np.sqrt(n_trials)
        
        # Store values
        late_time_means.append(late_time_mean_avg)
        late_time_stds.append(late_time_std_avg)
        late_time_means_err.append(late_time_mean_err)
        late_time_stds_err.append(late_time_std_err)
    
    # Plot data points with error bars for mean values
    ax1.errorbar(system_sizes, late_time_means, yerr=late_time_means_err,
                fmt='o', markersize=10, color=color, label=title,
                capsize=5, elinewidth=2, alpha=0.7)
    
    # Plot data points with error bars for std values
    ax2.errorbar(system_sizes, late_time_stds, yerr=late_time_stds_err,
                fmt='o', markersize=10, color=color, label=title,
                capsize=5, elinewidth=2, alpha=0.7)
    
    fit_results = {}
    
    # Fit exponential function to mean values
    try:
        # Convert to numpy arrays for fitting
        x_data = np.array(system_sizes)
        y_data = np.array(late_time_means)
        
        # Create smooth x values for plotting
        x_smooth = np.linspace(min(system_sizes) - 0.5, max(system_sizes) + 0.5, 100)
        
        # Initialize parameters for exponential fit
        p0 = [np.max(y_data) - np.min(y_data), 0.1, np.min(y_data)]
        bounds = ([0, -1, -np.inf], [np.inf, 1, np.inf])
        
        # Fit the data
        params, pcov = curve_fit(exp_func, x_data, y_data, p0=p0, maxfev=10000, bounds=bounds)
        a, b, c = params
        
        # Calculate parameter errors
        param_errors = np.sqrt(np.diag(pcov))
        
        # Calculate fit quality metrics
        y_pred = exp_func(x_data, *params)
        rss = np.sum((y_data - y_pred)**2)
        ss_tot = np.sum((y_data - np.mean(y_data))**2)
        r_squared = 1 - (rss / ss_tot)
        
        # Generate y values for smooth curve
        y_smooth = exp_func(x_smooth, *params)
        
        # Plot the exponential fit for mean
        ax1.plot(x_smooth, y_smooth, color=color, linewidth=3,
               label=f'Exp Fit: R={r_squared:.3f}')
        
        # Add text annotation for the fit parameters
        param_str = f"{a:.3f}e^({b:.3f}x) + {c:.3f}"
        ax1.annotate(f'Fit parameters:\n{param_str}\nR={r_squared:.3f}',
                   xy=(np.mean(system_sizes), max(y_data)),
                   xytext=(0, 20),
                   textcoords='offset points',
                   fontsize=12, color=color,
                   ha='center', va='center',
                   bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.8),
                   arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2', color=color))
        
        print(f"Mean - Exponential fit results:")
        print(f"A = {a:.6f}  {param_errors[0]:.6f}")
        print(f"B = {b:.6f}  {param_errors[1]:.6f}")
        print(f"C = {c:.6f}  {param_errors[2]:.6f}")
        print(f"R = {r_squared:.6f}")
        
        # Store mean fit results
        fit_results['mean'] = {
            'a': a,
            'b': b, 
            'c': c,
            'r_squared': r_squared,
            'xi': abs(1/b) if b != 0 else float('inf'),
            'xi_error': abs(param_errors[1]/(b**2)) if b != 0 else float('inf')
        }
        
    except Exception as e:
        print(f"Mean - Exponential fit failed: {e}")
        ax1.text(0.5, 0.5, f"Fit Failed: {str(e)}", transform=ax1.transAxes, 
               ha='center', va='center', fontsize=12, color='red')
    
    # Fit exponential function to standard deviation values
    try:
        # Convert to numpy arrays for fitting
        x_data = np.array(system_sizes)
        y_data = np.array(late_time_stds)
        
        # Create smooth x values for plotting
        x_smooth = np.linspace(min(system_sizes) - 0.5, max(system_sizes) + 0.5, 100)
        
        # Initialize parameters for exponential fit
        p0 = [np.max(y_data) - np.min(y_data), 0.1, np.min(y_data)]
        bounds = ([0, -1, -np.inf], [np.inf, 1, np.inf])
        
        # Fit the data
        params, pcov = curve_fit(exp_func, x_data, y_data, p0=p0, maxfev=10000, bounds=bounds)
        a, b, c = params
        
        # Calculate parameter errors
        param_errors = np.sqrt(np.diag(pcov))
        
        # Calculate fit quality metrics
        y_pred = exp_func(x_data, *params)
        rss = np.sum((y_data - y_pred)**2)
        ss_tot = np.sum((y_data - np.mean(y_data))**2)
        r_squared = 1 - (rss / ss_tot)
        
        # Generate y values for smooth curve
        y_smooth = exp_func(x_smooth, *params)
        
        # Plot the exponential fit for std
        ax2.plot(x_smooth, y_smooth, color=color, linewidth=3,
               label=f'Exp Fit: R={r_squared:.3f}')
        
        # Add text annotation for the fit parameters
        param_str = f"{a:.3f}e^({b:.3f}x) + {c:.3f}"
        ax2.annotate(f'Fit parameters:\n{param_str}\nR={r_squared:.3f}',
                   xy=(np.mean(system_sizes), max(y_data)),
                   xytext=(0, 20),
                   textcoords='offset points',
                   fontsize=12, color=color,
                   ha='center', va='center',
                   bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.8),
                   arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2', color=color))
        
        print(f"Std Dev - Exponential fit results:")
        print(f"A = {a:.6f}  {param_errors[0]:.6f}")
        print(f"B = {b:.6f}  {param_errors[1]:.6f}")
        print(f"C = {c:.6f}  {param_errors[2]:.6f}")
        print(f"R = {r_squared:.6f}")
        
        # Extract the characteristic length scale  from the exponential fit
        xi = abs(1/b) if b != 0 else float('inf')
        xi_error = abs(param_errors[1]/(b**2)) if b != 0 else float('inf')
        
        print(f"\nFinite Size Scaling Parameters:")
        print(f"Correlation length  = {xi:.6f}  {xi_error:.6f}")
        
        # Store std fit results
        fit_results['std'] = {
            'a': a,
            'b': b, 
            'c': c,
            'r_squared': r_squared,
            'xi': xi,
            'xi_error': xi_error
        }
        
        # --- FSS Analysis ---
        # Calculate scaling variable N/
        scaling_variable = x_data/xi
        
        # Calculate scaled observable ((N) - _)
        scaled_observable = y_data - c
        
        # Plot scaled data in FSS plot
        ax3.scatter(scaling_variable, scaled_observable, 
                  s=100, color=color, alpha=0.7)
        
        # Generate theoretical scaling function curve
        x_theory = np.linspace(0, max(scaling_variable)*1.2, 100)
        y_theory = a * np.exp(-x_theory)
        ax3.plot(x_theory, y_theory, 'k-', 
                linewidth=2, 
                label=f'Scaling function: f(N/) = {a:.3f}e^(-N/)')
        
        # Calculate data collapse quality
        predicted_scaled_values = a * np.exp(-scaling_variable)
        collapse_mse = np.mean((scaled_observable - predicted_scaled_values)**2)
        collapse_r2 = 1 - np.sum((scaled_observable - predicted_scaled_values)**2) / np.sum((scaled_observable - np.mean(scaled_observable))**2)
        
        print(f"Data collapse quality: MSE = {collapse_mse:.6f}, R = {collapse_r2:.6f}")
        
        # Determine if N is large enough to approach asymptotic behavior
        max_N_ratio = max(system_sizes) / xi
        if max_N_ratio > 3:
            approach_msg = f"The largest N value ({max(system_sizes)}) is {max_N_ratio:.1f}, suggesting the system is approaching asymptotic behavior."
        else:
            approach_msg = f"The largest N value ({max(system_sizes)}) is only {max_N_ratio:.1f}, suggesting the system may need larger N to reach asymptotic behavior."
            
        print(approach_msg)
        
        # Add annotation about FSS information
        ax3.annotate(f' = {xi:.2f}\n = {c:.4f}\n{approach_msg.split(",")[0]}',
                   xy=(0.05, 0.95), xycoords='axes fraction',
                   fontsize=12, 
                   va='top',
                   bbox=dict(boxstyle='round', fc='white', alpha=0.8))
        
        # Update std fit results
        fit_results['std'].update({
            'collapse_r2': collapse_r2,
            'max_N_ratio': max_N_ratio
        })
        
    except Exception as e:
        print(f"Std Dev - Exponential fit failed: {e}")
        ax2.text(0.5, 0.5, f"Fit Failed: {str(e)}", transform=ax2.transAxes, 
               ha='center', va='center', fontsize=12, color='red')
    
    # Configure mean plot
    ax1.grid(True, alpha=0.3)
    ax1.set_xlabel('System Size (N)', fontsize=14)
    ax1.set_ylabel('Late-Time Mean of Adjacency Values', fontsize=14)
    ax1.set_title('Late-Time Mean vs System Size', fontsize=16)
    ax1.set_xticks(system_sizes)
    
    # Configure std plot
    ax2.grid(True, alpha=0.3)
    ax2.set_xlabel('System Size (N)', fontsize=14)
    ax2.set_ylabel('Late-Time Std Dev of Adjacency Values', fontsize=14)
    ax2.set_title('Late-Time Std Dev vs System Size', fontsize=16)
    ax2.set_xticks(system_sizes)
    
    # Configure FSS plot
    ax3.grid(True, alpha=0.3)
    ax3.set_xlabel('Scaling Variable (N/)', fontsize=14)
    ax3.set_ylabel('Scaled Observable ((N) - )', fontsize=14)
    ax3.set_title('Finite Size Scaling Analysis', fontsize=16)
    
    # Use log scale for y-axis in FSS plot for exponential behavior
    ax3.set_yscale('log')
    ax3.legend(fontsize=12, loc='best')
    
    # Create simple legends that don't show all the fit labels
    for ax in [ax1, ax2]:
        handles, labels = ax.get_legend_handles_labels()
        # Get only the data points and final fit line
        if len(handles) >= 2:
            data_handle = handles[0]
            fit_handle = handles[-1]
            ax.legend([data_handle, fit_handle], [title, labels[-1]], fontsize=12, loc='upper left')
    
    plt.tight_layout()
    plt.savefig("adjacency_fss_analysis.png", dpi=300, bbox_inches='tight')
    plt.savefig("adjacency_fss_analysis.pdf", bbox_inches='tight')
    
    return fig, fit_results


# Example usage:
system_sizes = [6, 8, 10, 12]  # Your system sizes
fig, fit_results = plot_adjacency_with_fss(adj_matrices_datasets, system_sizes, title='Dataset 1')
plt.show()
```
