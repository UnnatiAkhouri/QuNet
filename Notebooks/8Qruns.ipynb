```python
threedms = get_3_qbit_dms(newdata1,8,"c2_2local","random")
threedms[0][0]

import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt

import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt

def von_neumann_entropy(rho):
    """
    Calculate the von Neumann entropy S = -Tr(rho log(rho)) for a density matrix.
    
    Args:
        rho: A numpy array representing a density matrix
        
    Returns:
        The von Neumann entropy as a real number
    """
    # Get the eigenvalues of the density matrix
    eigenvalues = linalg.eigvalsh(rho)
    
    # Remove very small eigenvalues that might cause numerical issues
    eigenvalues = eigenvalues[eigenvalues > 1e-10]
    
    # Calculate entropy: -sum(_i * log(_i))
    entropy = -np.sum(eigenvalues * np.log2(eigenvalues))
    
    return entropy

def analyze_entropy_per_step(dms, num_trials, num_steps=5, qubit_type="four"):
    """
    Calculate the mean and standard deviation of entropy across all subsystems
    for each step.
    
    Args:
        dms: The loaded density matrices (three or four qubit)
        num_trials: Number of trials to analyze
        num_steps: Number of steps to analyze (default is 5)
        qubit_type: Either "three" or "four" to indicate qubit type
        
    Returns:
        Dictionary with time steps as keys and (mean entropy, std entropy) as values
    """
    # Initialize results dictionary
    results = {}
    
    # Process each trial
    for trial_idx in range(num_trials):
        print(f"Processing trial {trial_idx+1}/{num_trials}")
        
        # Process each time step
        for step_idx in range(num_steps):
            # Get the density matrices for this step
            try:
                dms_at_step = dms[trial_idx][step_idx]
                
                # Calculate entropy for each subsystem
                entropies = []
                for qubits, dm in dms_at_step.items():
                    # Convert DM to numpy array
                    dm_array = dm.data.toarray()
                    
                    # Calculate entropy
                    try:
                        entropy = von_neumann_entropy(dm_array)
                        entropies.append(entropy)
                    except Exception as e:
                        print(f"Error calculating entropy for qubits {qubits} at step {step_idx}: {e}")
                
                # Calculate mean and std of entropies
                if entropies:  # If we have valid entropy values
                    mean_entropy = np.mean(entropies)
                    std_entropy = np.std(entropies)
                    
                    # Store results for this time step
                    if step_idx not in results:
                        results[step_idx] = []
                    
                    results[step_idx].append((mean_entropy, std_entropy))
            except (IndexError, KeyError) as e:
                print(f"Error accessing data for trial {trial_idx}, step {step_idx}: {e}")
                continue
    
    # Calculate final mean and std across all trials for each time step
    final_results = {}
    for step_idx, values in results.items():
        means = [v[0] for v in values]
        stds = [v[1] for v in values]
        
        # Mean of means, and mean of stds
        final_mean = np.mean(means)
        final_std = np.mean(stds)
        
        final_results[step_idx] = (final_mean, final_std)
    
    return final_results

def plot_entropy_results(results, qubit_type="four"):
    """
    Plot the entropy results.
    
    Args:
        results: Dictionary with step indices as keys and (mean entropy, std entropy) as values
        qubit_type: Either "three" or "four" to indicate qubit type
    """
    step_indices = sorted(results.keys())
    means = [results[step][0] for step in step_indices]
    stds = [results[step][1] for step in step_indices]
    
    plt.figure(figsize=(10, 6))
    plt.errorbar(step_indices, means, yerr=stds, marker='o', linestyle='-', capsize=5)
    plt.xlabel('Step Index')
    plt.ylabel('von Neumann Entropy (bits)')
    plt.title(f'Mean von Neumann Entropy of {qubit_type.capitalize()}-Qubit Subsystems')
    plt.grid(True)
    plt.savefig(f'{qubit_type}_qubit_entropy.png')
    plt.show()
    
    # Print the results
    print(f"{qubit_type.capitalize()}-Qubit Entropy Results:")
    print("Step Index | Mean Entropy | Standard Deviation")
    print("-" * 45)
    for step in step_indices:
        print(f"{step:10d} | {results[step][0]:12.6f} | {results[step][1]:12.6f}")

def analyze_qubit_entropy(dms, num_trials=10, num_steps=5, qubit_type="four"):
    """Main function to run the analysis"""
    
    # Calculate entropy statistics for each step
    results = analyze_entropy_per_step(dms, num_trials, num_steps, qubit_type)
    
    # Plot and print the results
    plot_entropy_results(results, qubit_type)
    
    return results

# Example usage for three-qubit DMs:
# entropy_3q = analyze_qubit_entropy(threedms, num_trials=len(threedms), qubit_type="three")

# Example usage for four-qubit DMs:
# entropy_4q = analyze_qubit_entropy(fourdms, num_trials=len(fourdms), qubit_type="four")

# Compare three and four qubit entropies
def compare_entropies(entropy_3q, entropy_4q):
    """
    Compare three and four qubit entropies with a side-by-side plot.
    
    Args:
        entropy_3q: Results from analyze_qubit_entropy for three-qubit systems
        entropy_4q: Results from analyze_qubit_entropy for four-qubit systems
    """
    steps = sorted(list(set(entropy_3q.keys()) & set(entropy_4q.keys())))
    
    means_3q = [entropy_3q[step][0] for step in steps]
    stds_3q = [entropy_3q[step][1] for step in steps]
    
    means_4q = [entropy_4q[step][0] for step in steps]
    stds_4q = [entropy_4q[step][1] for step in steps]
    
    plt.figure(figsize=(12, 7))
    
    plt.errorbar([s - 0.1 for s in steps], means_3q, yerr=stds_3q, 
                 marker='o', linestyle='-', capsize=5, label='Three-Qubit')
    plt.errorbar([s + 0.1 for s in steps], means_4q, yerr=stds_4q, 
                 marker='s', linestyle='-', capsize=5, label='Four-Qubit')
    
    plt.xlabel('Step Index')
    plt.ylabel('von Neumann Entropy (bits)')
    plt.title('Comparison of Three and Four-Qubit Entropies')
    plt.legend()
    plt.grid(True)
    plt.savefig('entropy_comparison.png')
    plt.show()

# Example usage:
threedms = get_3_qbit_dms(newdata1,8,"c2_2local","random")
fourdms = get_4_qbit_dms(newdata1,8,"c2_2local","random")

# Assuming threedms and fourdms are already loaded
num_trials_3q = len(threedms)
num_trials_4q = len(fourdms)
num_steps = 5  # Number of steps to analyze

# Calculate entropy for three-qubit systems
entropy_3q = analyze_qubit_entropy(threedms, num_trials_3q, num_steps, qubit_type="three")

# Calculate entropy for four-qubit systems
entropy_4q = analyze_qubit_entropy(fourdms, num_trials_4q, num_steps, qubit_type="four")

# Compare the results
compare_entropies(entropy_3q, entropy_4q)
```

```python
import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt
import matplotlib.cm as cm

def von_neumann_entropy(rho):
    """
    Calculate the von Neumann entropy S = -Tr(rho log(rho)) for a density matrix.
    
    Args:
        rho: A numpy array representing a density matrix
        
    Returns:
        The von Neumann entropy as a real number
    """
    # Get the eigenvalues of the density matrix
    eigenvalues = linalg.eigvalsh(rho)
    
    # Remove very small eigenvalues that might cause numerical issues
    eigenvalues = eigenvalues[eigenvalues > 1e-10]
    
    # Calculate entropy: -sum(_i * log(_i))
    entropy = -np.sum(eigenvalues * np.log2(eigenvalues))
    
    return entropy

def calculate_entropy_for_size(dms, trial_idx, step_idx, size):
    """
    Calculate entropies for all subsystems of a specific size in a given trial and step.
    
    Args:
        dms: Dictionary of density matrices (key: qubit_size -> trial_idx -> step_idx -> subsystem_tuple)
        trial_idx: Trial index
        step_idx: Step index
        size: Qubit subsystem size
        
    Returns:
        List of entropy values for all subsystems of the specified size
    """
    if size not in dms:
        return []
    
    try:
        subsystems = dms[size][trial_idx][step_idx]
        entropies = []
        
        for qubits, dm in subsystems.items():
            # Convert DM to numpy array
            dm_array = dm.data.toarray()
            
            # Calculate entropy
            try:
                entropy = von_neumann_entropy(dm_array)
                entropies.append(entropy)
            except Exception as e:
                print(f"Error calculating entropy for size {size}, qubits {qubits} at step {step_idx}: {e}")
        
        return entropies
    except (KeyError, IndexError) as e:
        print(f"Error accessing data for size {size}, trial {trial_idx}, step {step_idx}: {e}")
        return []

def analyze_entropy_across_sizes(dms_dict, num_trials, num_steps=5, max_size=8):
    """
    Calculate entropy statistics across different qubit subsystem sizes.
    
    Args:
        dms_dict: Dictionary of density matrices by size 
                 (e.g., {1: onedms, 2: twodms, 3: threedms, ...})
        num_trials: Number of trials
        num_steps: Number of steps to analyze
        max_size: Maximum qubit subsystem size
        
    Returns:
        Dictionary with structure: {step_idx: {size: (mean_entropy, std_entropy)}}
    """
    # Initialize results dictionary
    results = {step: {} for step in range(num_steps)}
    
    # Process each step
    for step_idx in range(num_steps):
        print(f"Processing step {step_idx+1}/{num_steps}")
        
        # Process each qubit subsystem size
        for size in range(1, max_size + 1):
            if size not in dms_dict:
                print(f"No data available for {size}-qubit subsystems")
                continue
                
            # Collect entropies across all trials for this step and size
            all_entropies = []
            
            for trial_idx in range(num_trials):
                entropies = calculate_entropy_for_size(dms_dict, trial_idx, step_idx, size)
                all_entropies.extend(entropies)
            
            # Calculate statistics if we have data
            if all_entropies:
                mean_entropy = np.mean(all_entropies)
                std_entropy = np.std(all_entropies)
                results[step_idx][size] = (mean_entropy, std_entropy)
            else:
                print(f"No entropy data for {size}-qubit subsystems at step {step_idx}")
    
    return results

def plot_entropy_by_size_and_step(entropy_results, num_steps=5, max_size=8,col="random"):
    """
    Plot entropy vs. subsystem size with different lines for each step.
    
    Args:
        entropy_results: Results from analyze_entropy_across_sizes
        num_steps: Number of steps that were analyzed
        max_size: Maximum qubit subsystem size
    """
    plt.figure(figsize=(12, 8))
    
    # Create a color map with different shades of blue
    if col=="random":
        blues = cm.Blues(np.linspace(0.3, 1.0, num_steps))
    if col=="mimic":
        blues = cm.Purples(np.linspace(0.3, 1.0, num_steps))
    if col =="greedy":
        blues = cm.Reds(np.linspace(0.3, 1.0, num_steps))
    # Plot each step with a different shade of blue
    for step_idx in range(num_steps):
        if step_idx not in entropy_results:
            continue
            
        sizes = sorted(entropy_results[step_idx].keys())
        means = [entropy_results[step_idx][size][0] for size in sizes]
        stds = [entropy_results[step_idx][size][1] for size in sizes]
        
        plt.errorbar(sizes, means, yerr=stds, 
                    marker='o', linestyle='-', capsize=5,
                    color=blues[step_idx], 
                    label=f'Step {step_idx}')
    
    plt.xlabel('Qubit Subsystem Size')
    plt.ylabel('von Neumann Entropy (bits)')
    plt.title('Entropy Scaling with Subsystem Size at Different Time Steps')
    plt.legend()
    plt.grid(True)
    plt.xticks(range(1, max_size + 1))
    plt.savefig('entropy_scaling.png')
    plt.show()
    
    # Print the results in a table
    print("Entropy Results by Size and Step:")
    print("Size | " + " | ".join([f"Step {i} MeanStd" for i in range(num_steps)]))
    print("-" * (7 + 19 * num_steps))
    
    for size in range(1, max_size + 1):
        row = [f"{size:4d}"]
        
        for step in range(num_steps):
            if step in entropy_results and size in entropy_results[step]:
                mean, std = entropy_results[step][size]
                row.append(f"{mean:.4f}{std:.4f}")
            else:
                row.append("N/A")
        
        print(" | ".join(row))

# Main function to run all analyses
def analyze_entropy_scaling(dms_dict, num_trials, num_steps=5, max_size=8,col="random"):
    """
    Analyze entropy scaling with subsystem size across different time steps.
    
    Args:
        dms_dict: Dictionary of density matrices by size
        num_trials: Number of trials
        num_steps: Number of steps
        max_size: Maximum qubit subsystem size
    """
    # Calculate entropy statistics for each size and step
    results = analyze_entropy_across_sizes(dms_dict, num_trials, num_steps, max_size)
    
    # Plot and print the results
    plot_entropy_by_size_and_step(results, num_steps, max_size,col)
    
    return results




twodms=get_2_qbit_dms(newdata1,8,"c2_2local","random")
threedms = get_3_qbit_dms(newdata1,8,"c2_2local","random")
fourdms = get_4_qbit_dms(newdata1,8,"c2_2local","random")
fivedms = get_5_qbit_dms(newdata1,8,"c2_2local","random")
sixdms = get_6_qbit_dms(newdata1,8,"c2_2local","random")

twodms_last5_all_trials=[]
for twodm_trial in twodms:
    twodms_last5_all_trials.append(twodm_trial[145:150])
len(twodms_last5_all_trials[0])

# Organize your data by qubit size
dms_dict = {
    2: twodms_last5_all_trials,
    3: threedms,    # One-qubit density matrices
    4: fourdms,    # Two-qubit density matrices
    5: fivedms,  # Three-qubit density matrices
    6: sixdms,   # Four-qubit density matrices
    # Add more sizes if available
}

# Set parameters
num_trials = min(len(dms) for dms in dms_dict.values())  # Use the smallest number of trials
num_steps = 5  # Number of steps (typically 5)
max_size = max(dms_dict.keys())  # Maximum size available

# Run the analysis
random_entropy_results = analyze_entropy_scaling(dms_dict, num_trials, num_steps, max_size,col="random")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
import matplotlib.lines as mlines

def plot_multiple_entropy_results(entropy_results_dict, title="Comparison of Entropy Across Simulations", 
                                 xlabel="Qubit Subsystem Size", ylabel="von Neumann Entropy (bits)",
                                 use_markers=True, use_errorbars=True,
                                 figsize=(12, 8), save_filename="entropy_comparison.png"):
    """
    Plot entropy results from multiple simulations on the same chart.
    
    Args:
        entropy_results_dict: Dictionary with structure {
            'simulation_name': {
                step_idx: {
                    size: (mean_entropy, std_entropy)
                }
            }
        }
        title: Title for the plot
        xlabel: Label for x-axis
        ylabel: Label for y-axis
        use_markers: Whether to use different markers for different steps
        use_errorbars: Whether to show error bars
        figsize: Size of the figure
        save_filename: Filename to save the plot
        
    Returns:
        matplotlib figure object
    """
    fig, ax = plt.subplots(figsize=figsize)
    
    # Define custom colormaps for each dataset
    custom_cmaps = [
        plt.cm.Blues,    # Shades of blue
        plt.cm.Purples,  # Shades of purple
        plt.cm.Reds,     # Shades of red
        plt.cm.Greens,   # Shades of green (if more than 3 datasets)
        plt.cm.Oranges,  # Shades of orange (if more than 4 datasets)
        plt.cm.YlOrBr,   # Yellow-Orange-Brown (if more than 5 datasets)
        plt.cm.GnBu,     # Green-Blue (if more than 6 datasets)
        plt.cm.PuRd      # Purple-Red (if more than 7 datasets)
    ]
    
    # Ensure we have enough colormaps for all datasets
    if len(entropy_results_dict) > len(custom_cmaps):
        print(f"Warning: Only {len(custom_cmaps)} colormaps defined, but {len(entropy_results_dict)} datasets provided.")
        print("Some datasets will use repeated colormaps.")
    
    # Define markers for different steps
    markers = ['o', 's', 'D', '^', 'v', '<', '>', 'p', '*', 'h']
    
    # Create legend handles
    simulation_handles = []
    step_handles = []
    
    # Find common sizes and steps across all simulations
    all_sizes = set()
    all_steps = set()
    
    for sim_name, sim_results in entropy_results_dict.items():
        for step, step_results in sim_results.items():
            all_steps.add(step)
            for size in step_results.keys():
                all_sizes.add(size)
    
    all_sizes = sorted(all_sizes)
    all_steps = sorted(all_steps)
    
    # Plot each simulation
    for sim_idx, (sim_name, sim_results) in enumerate(entropy_results_dict.items()):
        # Use the appropriate colormap for this dataset
        cmap_idx = sim_idx % len(custom_cmaps)
        cmap = custom_cmaps[cmap_idx]
        
        # Use a medium-dark color from this colormap for the simulation legend
        sim_color = cmap(0.7)  # 0.7 gives a medium-dark shade
        simulation_handles.append(mlines.Line2D([], [], color=sim_color, linestyle='-', 
                                               label=sim_name))
        
        # Get number of steps for this simulation
        sim_steps = sorted([step for step in sim_results.keys()])
        num_steps = len(sim_steps)
        
        # Plot each step with a different shade from the colormap
        for step_idx, step in enumerate(sim_steps):
            # Calculate the color intensity based on step index
            # Use 0.3 to 0.9 range to avoid too light or too dark colors
            color_intensity = 0.3 + 0.6 * (step_idx / max(1, num_steps - 1))
            step_color = cmap(color_intensity)
            
            sizes = []
            means = []
            stds = []
            
            # Collect data for all sizes in this step
            for size in all_sizes:
                if size in sim_results[step]:
                    mean, std = sim_results[step][size]
                    sizes.append(size)
                    means.append(mean)
                    stds.append(std)
            
            if not sizes:  # Skip if no data for this step
                continue
                
            # Use different marker for each step
            marker = markers[step_idx % len(markers)] if use_markers else None
            
            # Plot with or without error bars
            if use_errorbars:
                ax.errorbar(sizes, means, yerr=stds, marker=marker, linestyle='-',
                           color=step_color, capsize=3, 
                           label=f"{sim_name} - Step {step}")
            else:
                ax.plot(sizes, means, marker=marker, linestyle='-',
                       color=step_color, 
                       label=f"{sim_name} - Step {step}")
            
            # Add to step handles only once (assuming all simulations have the same steps)
            if sim_idx == 0:
                step_handles.append(mlines.Line2D([], [], marker=marker, color='black',
                                                 linestyle='None', label=f"Step {step}"))
    
    # Set labels and title
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.grid(True, alpha=0.3)
    
    # Set x-axis ticks to integer values
    ax.set_xticks(all_sizes)
    
    # Create two-part legend: one for simulations, one for steps
    first_legend = ax.legend(handles=simulation_handles, loc='upper left')
    ax.add_artist(first_legend)
    if use_markers:
        ax.legend(handles=step_handles, loc='lower right')
    
    plt.tight_layout()
    plt.savefig(save_filename, dpi=300, bbox_inches='tight')
    
    return fig

# Example usage:
def compare_multiple_simulations(sim_data_dict, num_trials=10, num_steps=5, max_size=8):
    """
    Compare entropy results from multiple simulations.
    
    Args:
        sim_data_dict: Dictionary with structure {
            'simulation_name': {
                size: density_matrices_data
            }
        }
        num_trials: Number of trials
        num_steps: Number of steps
        max_size: Maximum qubit subsystem size
        
    Returns:
        Combined entropy results
    """
    # Calculate entropy for each simulation
    combined_results = {}
    
    for sim_name, dms_dict in sim_data_dict.items():
        print(f"Analyzing simulation: {sim_name}")
        combined_results[sim_name] = analyze_entropy_across_sizes(
            dms_dict, num_trials, num_steps, max_size)
    
    # Plot the combined results
    plot_multiple_entropy_results(combined_results, 
                                title="Entropy Scaling Across Different Simulations",
                                save_filename="multi_sim_entropy_comparison.png")
    
    return combined_results


entropy_results_dict = {
    'Random (R1) Evolution': random_entropy_results,
    'Mimic (R5) Evolution': mimic_entropy_results,
    'Greedy (R2) Evolution': greedy_entropy_results
}

# Plot the combined results
plot_multiple_entropy_results(entropy_results_dict, 
                             title="Entropy Scaling Across Different Simulations",
                             save_filename="multi_sim_entropy_comparison.png")
```
